### Dog Breed Prediction CNN Model

#### Problem definition

This project aims to develop a CNN model that can accurately indentify and predict the breed of a dog.

The following steps are taken as follows in order to solve this problem:

1. Load the dataset from Kaggle.
2. Load labels from the CSV dataset.
3. Get the number of labels (breeds).
4. One-hot encode the breeds feature.
5. Load, convert into arrays and normalize the pictures.
6. Verify the shape and size of the X and y datasets.
7. Design the model network architechture.
8. Split the training set and fit the model with it and create an accuracy plot.
9. Evaluate the model for accuracy.
10. Predict wuth the test set images.

In this project, we will see how to use Keras and TensorFlow to build, train, and test a Convolutional Neural Network capable of identifying the breed of a dog in a supplied image. This is a supervised learning problem, specifically a multiclass classification problem.
